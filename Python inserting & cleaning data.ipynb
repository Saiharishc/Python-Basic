```python
df=pd.DataFrame(np.random.randn(5,3),index=['a0','a10','a20','a30','a40'],columns=['x','y','z'])
df
```

```python
df2=df.reindex(['a0','a10','a20','a28','a30','a40'])  #adding some extra row indexes, which will create null values in our DataFrame:

df2
```

```python
#If we want to replace null values with zero
df2.fillna(0)
```

```python
#If we want to fill null value with forward propagation, which means value previous to null value in the column will be used to fill null value
df2.fillna(method='pad')
```

```python
df2.fillna(df2.mean()) #fill with mean
```
